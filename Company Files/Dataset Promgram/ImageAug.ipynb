<h1 style = "font-weight:bold">데이터 증강기법 학습</h1>

> 데이터 증강기법에서 회전, 수평 및 수직 이동, 확대 및 축소, 반전, 잡음추가를 이미지에서 학습 시킬 수 있다.

```python
import random
import numpy as np
import os
import cv2
import glob
from PIL import Image
import PIL.ImageOps    
```

```python
# 다음 변수를 수정하여 새로 만들 이미지 갯수를 정함.
num_augmented_images = 50
```

윗 문단의 num_augmented_images는 새로 만들 이미지 개수를 정한다. <br>
만약 50장의 원본 사진이 폴더에 있고 num_augmentd_images를 30장으로 설정한다면, <br>
최종적으로 폴더에는 80장의 이미지가 저장된다. 



file_path에 원본 사진이 있는 폴더를 정해준다. <br>
file_names에는 위의 폴더 내부에 있는 이미지 이름의 배열이 저장된다. <br>
total_origin_image_num에는 file_names의 길이를 저장한다. <br>
만약 50장의 원본 이미지가 있다면, 이 변수는 50이 될 것이다. <br>
augment_cnt는 몇 번의 augmentation이 일어났는지를 확인하기 위해 선언함.

```python
file_path = 'C:\\Users\\kiot\\Yolo V5\\rotated\\'
file_names = os.listdir(file_path)
total_origin_image_num = len(file_names)
augment_cnt = 1

for i in range(1, num_augmented_images):
    change_picture_index = random.randrange(1, total_origin_image_num-1)
    print(change_picture_index)
    print(file_names[change_picture_index])
    file_name = file_names[change_picture_index]
    
    origin_image_path = file_path + file_name
    print(origin_image_path)
    image = Image.open(origin_image_path)
    random_augment = random.randrange(1,4)
    
    if(random_augment == 1):
        #이미지 좌우 반전
        print("invert")
        inverted_image = image.transpose(Image.FLIP_LEFT_RIGHT)
        inverted_image.save(file_path + 'inverted_' + str(augment_cnt) + '.png')
        
    elif(random_augment == 2):
        #이미지 기울이기
        print("rotate")
        rotated_image = image.rotate(random.randrange(-20, 20))
        rotated_image.save(file_path + 'rotated_' + str(augment_cnt) + '.png')
        
    elif(random_augment == 3):
        #노이즈 추가하기
        img = cv2.imread(origin_image_path)
        print("noise")
        row,col,ch= img.shape
        mean = 0
        var = 0.1
        sigma = var**0.5
        gauss = np.random.normal(mean,sigma,(row,col,ch))
        gauss = gauss.reshape(row,col,ch)
        noisy_array = img + gauss
        noisy_image = Image.fromarray(np.uint8(noisy_array)).convert('RGB')
        noisy_image.save(file_path + 'noiseAdded_' + str(augment_cnt) + '.png')
        
    augment_cnt += 1
```

for문의 첫 번째 파트는 원본 이미지와 Augmetation 방식을 모두 랜덤으로 설정하는 부분이다. <br>
change_picture_index로 전체 이미지 개수 중 하나를 랜덤하게 선택하여 <br>
file_name에 그 인덱스를 갖는 이미지 이름을 저장한다. 

origin_image_path에는 file_name에 저장해놓은 이름 덕분에 원본 이미지의 경로가 저장됨. <br>
image 변수에는 이미지가 탑재됨. <br>
random_augment 변수에는 1~3의 무작위 수가 결정됨. <br>
1이 선택 되었다면 : 원본 이미지를 좌우반전하여 원 폴더에 저장된다. <br>
2가 선택 되었다면 : 원본 이미지를 랜덤하게 -20 ~ 20도 회전하여 원 폴더에 저장한다. <br>
3이 선택 되었다면 : 원본 이미지에 가우스 노이즈를 추가하여 원 폴더에 저장한다.

노이즈 추가 부분은 가우스 함수에 따른 노이즈 추가 방식을 사용함 <br>
이미지를 불러와 가우스 함수에 따른 노이즈를 추가하고 원 이미지와 합치는 방식임 <br>
이미지를 합치고 난 뒤의 형식은 numpy.array 형식임. 따라서 바로 이미지로 저장이 되진 않는다. <br>

noisy_image = Image.fromarray(np.uint8(noisy_array)).convert('RGB') <br>
이 코드를 통해 array를 이미지로 바꾸어 원 폴더에 저장할 수 있음. <br>
augment_cnt는 한바퀴 돌 때마다 +1씩 업데이트 해줌.
